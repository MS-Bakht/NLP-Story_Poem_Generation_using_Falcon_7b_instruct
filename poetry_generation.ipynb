```python
!pip install transformers accelerate einops

from transformers import AutoTokenizer, AutoModelForCausalLM
import transformers
import torch

model = "tiiuae/falcon-7b-instruct" #"mrm8488/falcoder-7b" #tiiuae/falcon-40b-instruct

tokenizer = AutoTokenizer.from_pretrained(model)

pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
)
```

```python
theme = input('Enter theme: ')
# Example 'a father saying farewell to his sons before going to war'
```

```python
prompt = f"Generate a rhyming poem about '{theme}'. "

generation = pipeline(
    prompt,
    max_length=800,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
)
```

```python
output =''
for seq in generation:
    output = output + seq['generated_text']
```

```python
listing = output.split('\n')
del listing[0]
output = '\n'.join(listing)   # We need the points in form of array, to generate text

output
```

```python
print(output)
```
